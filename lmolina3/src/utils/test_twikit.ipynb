```python
from twikit import Client
from account_info import USERNAME, EMAIL, PASSWORD

# Initialize client
client = Client('en-US')

client.login(
    auth_info_1=USERNAME ,
    auth_info_2=EMAIL,
    password=PASSWORD
)

# save cookies in order to pull data without getting banned
client.save_cookies('cookies.json')
```

```python
# loads the cookies that were saved
client.load_cookies('cookies.json')

# this gets the users id 
user = client.get_user_by_screen_name(screen_name='MrBeast')
print(user.id)
print(user.name)
print(user.screen_name)
print(user.profile_banner_url)
print(user.profile_image_url)
print(user.description)
print(user.followers_count)
print(user.following_count)


# gets the first 20 tweets of the user and prints them
tweets = client.get_user_tweets(user.id,'Tweets', count=20)
for tweet in tweets:
    print(tweet)
    print(tweet.text)
    print(tweet.get_favoriters)
    for favorites in client.get_favoriters(tweet.id,40):
        print('user_'+ favorites.id)
    print(tweet.get_retweeters)
    print('comment_'+tweet.id)
```

```python
import pandas as pd
import json

# file path to excel sheet
file_path = r"C:\Users\16822\Research Project SER\SEP-NHANES\lmolina3\src\utils\codes.xlsx"

# reads in excel sheet into pandas data frame
df = pd.read_excel(file_path)


# this turns the street name and its code name into list 
street_name_list = df['Street Name'].astype(str).values.tolist()
street_name_code_list = df['Street Name Code'].astype(str).values.tolist()

# this dictionary contains the keyname in the keys and keyname id in values
keyname_id = {}
for i in range(len(street_name_list)):
    keyname_id[street_name_list[i]] = street_name_code_list[i]
print(json.dumps(keyname_id, indent =4))
```

```python
import urllib.request
import os

def download_image(URL, save_as):
    urllib.request.urlretrieve(URL,save_as)

def delete_jpg(file):
    if os.path.exists(file):
        os.remove(file)
    else:
        print('File not found')
```
