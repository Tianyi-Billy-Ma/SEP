```python
from twikit import Client
from account_info import USERNAME, EMAIL, PASSWORD

# Initialize client
client = Client('en-US')

client.login(
    auth_info_1=USERNAME ,
    auth_info_2=EMAIL,
    password=PASSWORD
)

# save cookies in order to pull data without getting banned
client.save_cookies('cookies.json')
```

```python
# loads the cookies that were saved
client.load_cookies('cookies.json')

# this gets the users id 
user = client.get_user_by_screen_name(screen_name='MrBeast')
print(user.id)
print(user.name)
print(user.screen_name)
print(user.profile_banner_url)
print(user.profile_image_url)
print(user.description)
print(user.followers_count)
print(user.following_count)


# gets the first 20 tweets of the user and prints them
tweets = client.get_user_tweets(user.id,'Tweets', count=20)
for tweet in tweets:
    print(tweet)
    print(tweet.text)
    print(tweet.get_favoriters)
    print(tweet.get_retweeters)
```
